# 인식 결과 비교

```python
import numpy as np
import cv2
import random
import matplotlib.pyplot as plt
from PIL import ImageFont, ImageDraw, Image
import easyocr
import json

# easyocr 인식 실행
reader = easyocr.Reader(['ko','en'], gpu=True)
result = reader.readtext("./5350109-2001-0001-0001.jpg", width_ths=0.1, adjust_contrast=0.1)

# 인식 결과 출력
print("<OCR 인식 결과>")
for i in result:
    print(i[1])
    
print("")
print("####################################################################################################")
print("")

# 라벨링 데이터(json) 파일 불러오기
with open("./5350109-2001-0001-0001.json", encoding="UTF8") as file:
    data = json.load(file)
    json_test = data['annotations']
    
    # 문자 값만 출력
    print("<라벨링 데이터>")
    for k in json_test:
        print(k['annotation.text'])
    
    # 인식 결과값과 라벨링 데이터 값 추출
    result_texts = {i[1] for i in result}
    json_test_texts = {k['annotation.text'] for k in json_test}
    
    # 일치 비율 계산
    matching_texts = result_texts.intersection(json_test_texts)
    percentage = (len(matching_texts) / len(result_texts)) * 100
    
    print("")
    print("####################################################################################################")
    print("")
    
    # 일치율 및 정확도 출력
    if matching_texts:
        print("인식 성공 : ", matching_texts)
        print("")
        print("####################################################################################################")
        print("")
        print("정확도 : {:.2f}%".format(percentage))
    else:
        print("정확한 인식 값이 없습니다.")
```

순서 상관없이 단순 배열 값 비교여서 정확도 계산에서 오차가 존재하는 것으로 예상

# 인식 결과 비교 (라벨링 데이터 전부) - 수정중

```python
import numpy as np
import cv2
import random
import matplotlib.pyplot as plt
from PIL import ImageFont, ImageDraw, Image
import easyocr
import json
import os

# 파일 경로
folder_path = '2001'
# 폴더 안에 있는 모든 파일을 리스트 형태로 가져오기
file_list = os.listdir(folder_path)
# 파일 경로와 파일 이름 결합
file_paths = [os.path.join(folder_path, file_name) for file_name in file_list]

# 파일 5개만 선택 (해당 코드 삭제 시 모든 파일 실행)
file_paths = file_paths[:4]

# easyocr 모델 로드
reader = easyocr.Reader(['ko','en'], gpu=True)

# 각 파일에 대해 OCR 수행
for file_path in file_paths:
    print("파일명 :", os.path.basename(file_path))
    result = reader.readtext(file_path, width_ths=0.1, adjust_contrast=0.1)
    
    for i in result:
        print(i[1])
        
    print("")
    print("####################################################################################################")
    print("")
    
    # 라벨링 데이터(json) 파일 불러오기
    with open("./5350109-2001-0001-0001.json", encoding="UTF8") as file:
        data = json.load(file)
        json_test = data['annotations']
    
        # 문자 값만 출력
        print("<라벨링 데이터>")
        for k in json_test:
            print(k['annotation.text'])
    
        # 인식 결과값과 라벨링 데이터 값 추출
        result_texts = {i[1] for i in result}
        json_test_texts = {k['annotation.text'] for k in json_test}
    
        # 일치 비율 계산
        matching_texts = result_texts.intersection(json_test_texts)
        percentage = (len(matching_texts) / len(result_texts)) * 100
    
        print("")
        print("####################################################################################################")
        print("")
    
        # 일치율 및 정확도 출력
        if matching_texts:
            print("인식 성공 : ", matching_texts)
            print("")
            print("####################################################################################################")
            print("")
            print("정확도 : {:.2f}%".format(percentage))
        else:
            print("정확한 인식 값이 없습니다.")
```
