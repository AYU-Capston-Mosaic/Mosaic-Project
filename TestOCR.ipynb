# 인식 결과 비교

```python
import numpy as np
import cv2
import random
import matplotlib.pyplot as plt
from PIL import ImageFont, ImageDraw, Image
import easyocr
import json

# easyocr 인식 실행
reader = easyocr.Reader(['ko','en'], gpu=True)
result = reader.readtext("./5350109-2001-0001-0001.jpg", width_ths=0.1, adjust_contrast=0.1)

# 인식 결과 출력
print("<OCR 인식 결과>")
for i in result:
    print(i[1])
    
print("")
print("####################################################################################################")
print("")

# 라벨링 데이터(json) 파일 불러오기
with open("./5350109-2001-0001-0001.json", encoding="UTF8") as file:
    data = json.load(file)
    json_test = data['annotations']
    
    # 문자 값만 출력
    print("<라벨링 데이터>")
    for k in json_test:
        print(k['annotation.text'])
    
    # 인식 결과값과 라벨링 데이터 값 추출
    result_texts = {i[1] for i in result}
    json_test_texts = {k['annotation.text'] for k in json_test}
    
    # 일치 비율 계산
    matching_texts = result_texts.intersection(json_test_texts)
    percentage = (len(matching_texts) / len(result_texts)) * 100
    
    print("")
    print("####################################################################################################")
    print("")
    
    # 일치율 및 정확도 출력
    if matching_texts:
        print("인식 성공 : ", matching_texts)
        print("")
        print("####################################################################################################")
        print("")
        print("정확도 : {:.2f}%".format(percentage))
    else:
        print("정확한 인식 값이 없습니다.")
```

# 인식 결과 비교 (5개의 문서만 실행)

```python
import numpy as np
import cv2
import random
import matplotlib.pyplot as plt
from PIL import ImageFont, ImageDraw, Image
import easyocr
import json
import os
import time

# 폴더 경로
folder_path_source = '2001_source'
folder_path_label = '2001_label'

# 폴더 안에 있는 모든 파일을 리스트 형태로 가져오기
file_list = os.listdir(folder_path_source)

# 파일 경로와 파일 이름 결합
file_paths = [os.path.join(folder_path_source, file_name) for file_name in file_list]

# 인식 시간 고려하여 파일 5개만 선택하여 실행
file_paths = file_paths[:5]

# easyocr 모델 로드
reader = easyocr.Reader(['ko','en'], gpu=True)

# 파일 수와 평균 경과 시간, 평균 정확도를 저장할 변수 초기화
file_count = len(file_paths)
total_elapsed_time = 0
total_accuracy = 0

# 각 파일에 대해 OCR 수행
for file_path in file_paths:
    filename = os.path.splitext(os.path.basename(file_path))[0]
    print("파일명 :", os.path.basename(file_path))
    
    # 시작 시간 저장
    start_time = time.time()
    
    # OCR 인식 결과 추출
    result = reader.readtext(folder_path_source+"/"+filename+".jpg", width_ths=0.1, adjust_contrast=0.1)
    
    # 인식 경과 시간 계산
    elapsed_time = time.time() - start_time
    
    # 총 경과 시간 계산
    total_elapsed_time += elapsed_time
    
    # 문자 인식 결과값 출력
    print("<OCR 인식 결과>")
    for i in result:
        print(i[1])

    print("####################################################################################################")
    
    # 라벨링 데이터(json) 파일 불러오기
    with open(folder_path_label+"/"+filename+".json", encoding="UTF8") as file:
        data = json.load(file)
        json_test = data['annotations']
    
        # 문자 값만 출력
        print("<라벨링 데이터>")
        for k in json_test:
            print(k['annotation.text'])
    
        # 인식 결과값과 라벨링 데이터 값 추출
        result_texts = {i[1] for i in result}
        json_test_texts = {k['annotation.text'] for k in json_test}
    
        # 일치 비율 계산
        matching_texts = result_texts.intersection(json_test_texts)
        percentage = (len(matching_texts) / len(result_texts)) * 100
    
        print("####################################################################################################")
    
        # 일치율 및 정확도, 경과 시간 출력
        if matching_texts:
            print("인식 성공 : ", matching_texts)
            print("####################################################################################################")
            print("정확도 : {:.2f}%".format(percentage))
            print("경과 시간 : {:.3f}초".format(elapsed_time))
            total_accuracy += percentage
        else:
            print("정확한 인식 값이 없습니다.")
            
        print("")
        print("####################################################################################################")
        print("####################################################################################################")
        print("####################################################################################################")
        print("")
        
# 평균 경과 시간과 평균 정확도 계산 및 출력
average_elapsed_time = total_elapsed_time / file_count
average_accuracy = total_accuracy / file_count
print("평균 경과 시간 : {:.3f}초".format(average_elapsed_time))
print("평균 정확도 : {:.2f}%".format(average_accuracy))
```
