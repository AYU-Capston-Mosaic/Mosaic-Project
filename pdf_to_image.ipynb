```python
from pdf2image import convert_from_path

file_name = "simpledoc.pdf" # pdf 파일 이름

pages = convert_from_path("./" + file_name) # 파일 경로

for i, page in enumerate(pages):
	page.save("./source/"+file_name+str(i)+".jpg", "JPEG")
```
