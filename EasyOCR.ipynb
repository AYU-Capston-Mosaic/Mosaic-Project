## 기본 인식 및 출력

```python
import easyocr
 
reader = easyocr.Reader(['ko','en'])    # 인식할 언어 리스트(한국어, 영어) 지정
result = reader.readtext('EasyOCR_Test_Image.jpg')    # 인식할 이미지(상대경로값) 지정

print(result)
```

easyocr 모듈로 간단하게 이미지 내의 문자를 인식하여 출력하는 코드이다.
Reader 함수로 인식할 언어를 한국어와 영어로 지정하여 리스트 생성하고
readtext 함수로 인식할 이미지의 경로를 입력해 문자 인식을 실행한다.
최종적으로 print 함수로 출력하면 지정한 언어의 문자 뿐만 아니라
추출된 영역의 좌표, 크기 등의 정보까지 출력된다.

추가로 Reader 함수에 gpu 속성으로 GPU로 모듈을 실행할 수 있다. (ex) gpu = True / 기본값은 CPU이며 False로 직접 지정할 수도 있다.)

## 이미지에 인식 영역 및 인식 문자 출력 (검색 자료)

```python
import easyocr
import numpy as np
import cv2
import random
import matplotlib.pyplot as plt
from PIL import ImageFont, ImageDraw, Image

reader = easyocr.Reader(['ko','en'])    # 인식할 언어 리스트(한국어, 영어) 지정
result = reader.readtext('EasyOCR_Test_Image.jpg')    # 인식할 이미지(상대경로값) 지정(인식 영역 출력으로 인해 detail 요소 미삭제)

img = cv2.imread('EasyOCR_Test_Image.jpg')

img = Image.fromarray(img)    # 이미지에 그리기 위해서 CV2 이미지를 PIL 이미지 객체로 변환
font = ImageFont.truetype('gulim.ttc',15)
draw = ImageDraw.Draw(img)

np.random.seed(42)
COLORS = np.random.randint(0, 255, size=(255, 3),dtype="uint8")

for i in result :
    # 이미지 좌표 변수 지정
    x = i[0][0][0] 
    y = i[0][0][1] 
    w = i[0][1][0] - i[0][0][0] 
    h = i[0][2][1] - i[0][1][1]
    
    # 랜덤 색상 지정
    color_idx = random.randint(0,255)
    color = [int(c) for c in COLORS[color_idx]]

    # 이미지에 인식 영역 / 인식 문자 출력
    draw.rectangle(((x, y), (x+w, y+h)), outline=tuple(color), width=2)
    draw.text((int((x + x + w) / 2) , y-2),str(i[1]), font=font, fill=tuple(color),)

plt.figure(figsize=(50,50))
plt.imshow(img)
plt.show()
```

## 인식 영역(번호 지정)과 문자 인식 결과만 출력

```python
import easyocr
import numpy as np
import cv2
import matplotlib.pyplot as plt

file_path = 'EasyOCR_Test_Image.jpg'

reader = easyocr.Reader(['ko','en'])
result = reader.readtext(file_path)
result_text = reader.readtext(file_path, detail=0)

img = cv2.imread(file_path)
img = Image.fromarray(img)
draw = ImageDraw.Draw(img)

count = 0

for i in result :
    count+=1
    x = i[0][0][0] 
    y = i[0][0][1] 
    w = i[0][1][0] - i[0][0][0] 
    h = i[0][2][1] - i[0][1][1]

    draw.rectangle(((x, y), (x+w, y+h)), width=2)
    draw.text((int((x + x + w) / 2) , y-2), str(count))

plt.imshow(img)
plt.show()

print(result_text)
```

본 프로젝트에 필요할 것으로 예상되는 요소만 출력하도록 변경하였다.

출력 요소
    1. 문자 인식 영역(영역별로 번호 지정)
    2. 인식 결과(인식한 문자만 출력)

제거 요소
    1. 색상(랜덤 요소 포함), 폰트, 출력 이미지 크기 지정
    2. 인식한 내용을 이미지에 출력
    
변경 및 보완해야 될 점
    1. 경로 지정으로 이미지 불러오는 방식 변경
    2. 이미지에 영역 그리는 것으로 인해 detail 요소 사용 불가. 때문에 문자 인식 변수(result)를 2번 사용
    3. 한글 문자 인식률 저조

## GUI 기반 이미지 불러오기 및 문자 인식

```python
import easyocr
import numpy as np
import cv2
import matplotlib.pyplot as plt

from tkinter import *
from tkinter import filedialog
from tkinter.simpledialog import *
import tkinter.messagebox as msgbox
from PIL import ImageFont, ImageDraw, Image, ImageTk

def main() :
    
    win = Tk()
    win.title("EasyOCR 문자 인식")
    win.geometry('400x300')

    label = Label(win, text="EasyOCR 문자 인식", font=("맑은 고딕",25), height=2)
    label.pack()
    
    btn = Button(win, text="이미지 업로드", height = 10, width = 25, command = text_ocr())
    btn.place(x = 100, y = 100)

    win.mainloop()
    
def text_ocr() :
    
    file_path = filedialog.askopenfilename(initialdir = "", title = "파일 열기",
                                          filetypes = (("모든 파일", "*.*"), ("jpg 파일", '*.jpg'), ("png 파일", '*.png')))
    
    reader = easyocr.Reader(['ko','en'])
    result = reader.readtext(file_path)
    result_text = reader.readtext(file_path, detail=0)
    
    img = cv2.imread(file_path)
    img = Image.fromarray(img)
    draw = ImageDraw.Draw(img)

    count = 0

    for i in result :
        count+=1
        x = i[0][0][0]
        y = i[0][0][1]
        w = i[0][1][0] - i[0][0][0]
        h = i[0][2][1] - i[0][1][1]

        draw.rectangle(((x, y), (x+w, y+h)), width=2)
        draw.text((int((x + x + w) / 2) , y-2), str(count))

    plt.imshow(img)
    plt.show()

    print(result_text)
    
# ------------------------------ 실행 ------------------------------

main()

cv2.waitKey()
cv2.destroyAllWindows()
```

문제점
    1. 한글 경로 오류
    2. 실행 시 버튼 누르지도 않았는데 바로 파일 선택 창 실행됨
    3. 다중 실행 불가 (한번 실행하고 종료해야 됨)
    4. 문자 인식률 저조
