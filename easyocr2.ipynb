```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

src = cv2.imread('C:/Users/82109/Desktop/12.jpg', cv2.IMREAD_GRAYSCALE)

alpha1 = -0.5 # 명암 조정값
alpha2 = 1.0 # 명암 조정값

dst1 = np.clip((1+alpha1) * src - 128 * alpha1, 0, 255).astype(np.uint8)

dst2 = np.clip((1+alpha2) * src - 128 * alpha2, 0, 255).astype(np.uint8)
imgs = {'or' : src, '1':dst1, '2':dst2}
for i ,(k,v) in enumerate(imgs.items()):
    plt.subplot(2,2,i+1)
    plt.title(k)
    plt.imshow(v,'gray')
    plt.xticks([]), plt.yticks([])
```
